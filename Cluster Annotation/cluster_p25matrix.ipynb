```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
psmfiles = os.listdir('PSMTables')

print(psmfiles)

#merging the files into 1 
all_psm = []
for file in psmfiles:
    df = (
        pd.read_csv(f'PSMTables/{file}',
                     delimiter="\t")
                     .assign(File = file.split("_")[0]) #function to add a column at the end with a shorter name
    )
    all_psm.append(df) #this is to add files together

def get_spectrum(row): #function to rename the spectrum id in the psm file dataframe
    spectra_raw = row.Spectrum
    return spectra_raw.split("_")[-1]

main_psm = (pd.concat(all_psm).assign(Spectrum_ID = lambda df: df.apply(get_spectrum, axis=1))
)

main_psm['File'] = main_psm['File'].astype(str)
main_psm['Spectrum_ID'] = main_psm['Spectrum_ID'].astype(int)

main_psm #main working psm df
```

```python
p25files = os.listdir('C:/Users/Izadora/Documents/Project2/spectralclustering/p25_clusterfiles/rawclusters')

all_p25 = []

for file in p25files:
    df = (
        pd.read_csv(f'C:/Users/Izadora/Documents/Project2/spectralclustering/p25_clusterfiles/rawclusters/{file}', index_col=0)
    )
    all_p25.append(df)



filenames = {'AB15.mgf':'AB15',
             'AB16.mgf':'AB16',
             'AB18.mgf':'AB18',
             'AB19.mgf':'AB19',
             'ZK01.mgf':'ZK01',
             'MM01.mgf':'MM01'}

allp25 = (
    pd.concat(all_p25)
    .sort_values('Number of Files', ascending =False)
    .rename(columns = {'Number of Files':'FileCount', 'Filename':'File'})
    .replace({'File': filenames})

)


allp25['File'] = allp25['File'].astype(str)
allp25['Spectrum_ID'] = allp25['Spectrum_ID'].astype(int)

#this will be the working maracluster df
allp25
```

```python
#mapping
p25_map = pd.merge(
    main_psm,
    allp25,
    how = 'right',
    on = ['File','Spectrum_ID']
)

#now we use this for the subcategorisation
p25_map
```

```python
#load the PEAKS file, to compare with the psm mapped file, only using the alldeepnovo file, not including those from the reference databases

peaks = pd.read_csv(
    'C:/Users/Izadora/Documents/Project2/spectralclustering/PEAKS/alldeepnovo.csv',
    index_col=0
)

#just to differentiate between PSM peptides with PEAKS peptide
peaks1 = peaks.rename(columns={'Peptide':'Peaks_Peptide'})

peaks1
```

```python
#merge the peaks with the p25_map to have the peptides in the same file

merge = pd.merge(
    p25_map,
    peaks1,
    how = 'left',
    on = ['File','Spectrum_ID']
)   

merge.to_csv('allp25_matrix/all_p25_PSMPEAKS.csv')
```

```python
merge #now we use this to do the clustering
```

```python
merge.columns
```

```python
#clean the df so that only the relevent info is there

clean = merge[['Cluster_ID','Cluster_Size','Peptide','Type','PEPscore',
                'FDR','Delta','Conflict','Annotation', 'Database','Charge','Proteins',
                'Peaks_Peptide','SCORE (%)','ALC (%)','m/z', 'z', 'RT', 'Delta RT',
                'FileCount','Spectrum_ID','File',]]

clean
```

```python
#let's check how many clusters and novel peptides found there are in these files

cluster_count = clean['Cluster_ID'].nunique()

print('Number of filtered clusters for p25 in 2-6 files:', cluster_count)


#check the number of peptides in this clusters
peptide_count = clean[['Cluster_ID','Peptide','Type']].drop_duplicates()

peptide_count = peptide_count['Type'].value_counts()

print('Type of peptides:',peptide_count)
```

```python
#number of clusters for each file

cluster2 = clean.loc[clean['FileCount']==2].groupby(by='Cluster_ID')
cluster3 = clean.loc[clean['FileCount']==3].groupby(by='Cluster_ID')
cluster4 = clean.loc[clean['FileCount']==4].groupby(by='Cluster_ID')
cluster5 = clean.loc[clean['FileCount']==5].groupby(by='Cluster_ID')
cluster6 = clean.loc[clean['FileCount']==6].groupby(by='Cluster_ID')

print (len(cluster2)), 
print (len(cluster3)), 
print (len(cluster4)), 
print (len(cluster5)), 
print (len(cluster6))
```

```python
#make into different clusters for ease of mapping later
cluster2 = clean.loc[clean['FileCount']==2]
cluster3 = clean.loc[clean['FileCount']==3]
cluster4 = clean.loc[clean['FileCount']==4]
cluster5 = clean.loc[clean['FileCount']==5]
cluster6 = clean.loc[clean['FileCount']==6]
```

```python
#categorise into 2: fully unidentified and identified

group = clean.groupby('Cluster_ID').count().reset_index()

group
```

```python
#for fully unidentified, we can say those with 0 peptides

unid = group.loc[group['Peptide']==0]

unid_clusters = unid[['Cluster_ID']]

unid_clusters #now use this to map onto the clusters and check how many there are
```

```python
#map for fully uniden

A = unid_clusters
B = cluster2
C = cluster3
D = cluster4
E = cluster5
F = cluster6

unid_2 = A[A.Cluster_ID.isin(B.Cluster_ID)]
unid_3 = A[A.Cluster_ID.isin(C.Cluster_ID)]
unid_4 = A[A.Cluster_ID.isin(D.Cluster_ID)]
unid_5 = A[A.Cluster_ID.isin(E.Cluster_ID)]
unid_6 = A[A.Cluster_ID.isin(F.Cluster_ID)]


print('Unidentified (2) clusters=', len(unid_2))
print('Unidentified (3) clusters=', len(unid_3))
print('Unidentified (4) clusters=', len(unid_4))
print('Unidentified (5) clusters=', len(unid_5))
print('Unidentified (6) clusters=', len(unid_6))
print('total unidentified clusters in p25 = ', len(unid_2)+len(unid_3)+len(unid_4)+len(unid_5)+len(unid_6))
```

```python
#merge

unid_file2 = pd.merge(
    cluster2,
    unid_2,
    how = 'right',
    on = 'Cluster_ID'
)
unid_file3 = pd.merge(
    cluster3,
    unid_3,
    how = 'right',
    on = 'Cluster_ID'
)
unid_file4 = pd.merge(
    cluster4,
    unid_4,
    how = 'right',
    on = 'Cluster_ID'
)
unid_file5 = pd.merge(
    cluster5,
    unid_5,
    how = 'right',
    on = 'Cluster_ID'
)
unid_file6 = pd.merge(
    cluster6,
    unid_6,
    how = 'right',
    on = 'Cluster_ID'
)


#save these files
unid_file2.to_excel('allp25_matrix/fully_unid_clust(2).xlsx')
unid_file3.to_excel('allp25_matrix/fully_unid_clust(3).xlsx')
unid_file4.to_excel('allp25_matrix/fully_unid_clust(4).xlsx')
unid_file5.to_excel('allp25_matrix/fully_unid_clust(5).xlsx')
unid_file6.to_excel('allp25_matrix/fully_unid_clust(6).xlsx')
```

```python
#filter out all the unidentified from the cluster df, this will be used from now on

filter_unid2 = cluster2[~cluster2.Cluster_ID.isin(unid_file2.Cluster_ID)]
filter_unid3 = cluster3[~cluster3.Cluster_ID.isin(unid_file3.Cluster_ID)]
filter_unid4 = cluster4[~cluster4.Cluster_ID.isin(unid_file4.Cluster_ID)]
filter_unid5 = cluster5[~cluster5.Cluster_ID.isin(unid_file5.Cluster_ID)]
filter_unid6 = cluster6[~cluster6.Cluster_ID.isin(unid_file6.Cluster_ID)]
```

```python
#now we wanna filter into peptide counts
pep_fil2 = filter_unid2.groupby(by='Cluster_ID').nunique().reset_index()
pep_fil3 = filter_unid3.groupby(by='Cluster_ID').nunique().reset_index()
pep_fil4 = filter_unid4.groupby(by='Cluster_ID').nunique().reset_index()
pep_fil5 = filter_unid5.groupby(by='Cluster_ID').nunique().reset_index()
pep_fil6 = filter_unid6.groupby(by='Cluster_ID').nunique().reset_index()

#specify the identified/partial peptide clusters list
id_clust2 = pep_fil2.loc[pep_fil2['Peptide']==1]
id_clust3 = pep_fil3.loc[pep_fil3['Peptide']==1]
id_clust4 = pep_fil4.loc[pep_fil4['Peptide']==1]
id_clust5 = pep_fil5.loc[pep_fil5['Peptide']==1]
id_clust6 = pep_fil6.loc[pep_fil6['Peptide']==1]

#extract out all of the id and par clusters
par_id2 = filter_unid2[filter_unid2.Cluster_ID.isin(id_clust2.Cluster_ID)]
par_id3 = filter_unid3[filter_unid3.Cluster_ID.isin(id_clust3.Cluster_ID)]
par_id4 = filter_unid4[filter_unid4.Cluster_ID.isin(id_clust4.Cluster_ID)]
par_id5 = filter_unid5[filter_unid5.Cluster_ID.isin(id_clust5.Cluster_ID)]
par_id6 = filter_unid6[filter_unid6.Cluster_ID.isin(id_clust6.Cluster_ID)]
```

```python
#extract only the partial clusters!
par2 = par_id2[par_id2['Peptide'].isna()]
par3 = par_id3[par_id3['Peptide'].isna()]
par4 = par_id4[par_id4['Peptide'].isna()]
par5 = par_id5[par_id5['Peptide'].isna()]
par6 = par_id6[par_id6['Peptide'].isna()]


print('Partial (2) clusters=', len(par2.groupby('Cluster_ID')))
print('Partial (3) clusters=', len(par3.groupby('Cluster_ID')))
print('Partial (4) clusters=', len(par4.groupby('Cluster_ID')))
print('Partial (5) clusters=', len(par5.groupby('Cluster_ID')))
print('Partial (6) clusters=', len(par6.groupby('Cluster_ID')))
```

```python
#now merge with the original cluster df

p2 = par_id2[par_id2.Cluster_ID.isin(par2.Cluster_ID)]
p3 = par_id3[par_id3.Cluster_ID.isin(par3.Cluster_ID)]
p4 = par_id4[par_id4.Cluster_ID.isin(par4.Cluster_ID)]
p5 = par_id5[par_id5.Cluster_ID.isin(par5.Cluster_ID)]
p6 = par_id6[par_id6.Cluster_ID.isin(par6.Cluster_ID)]
```

```python
#save the files
p2.to_excel('allp25_matrix/partial_clust(2).xlsx')
p3.to_excel('allp25_matrix/partial_clust(3).xlsx')
p4.to_excel('allp25_matrix/partial_clust(4).xlsx')
p5.to_excel('allp25_matrix/partial_clust(5).xlsx')
p6.to_excel('allp25_matrix/partial_clust(6).xlsx')
```

```python
#for the fully identified, we can filter out from the p2-p6

id2 = par_id2[~par_id2.Cluster_ID.isin(par2.Cluster_ID)]
id3 = par_id3[~par_id3.Cluster_ID.isin(par3.Cluster_ID)]
id4 = par_id4[~par_id4.Cluster_ID.isin(par4.Cluster_ID)]
id5 = par_id5[~par_id5.Cluster_ID.isin(par5.Cluster_ID)]
id6 = par_id6[~par_id6.Cluster_ID.isin(par6.Cluster_ID)]


print('Identified (2) clusters=', len(id2.groupby('Cluster_ID')))
print('Identified (3) clusters=', len(id3.groupby('Cluster_ID')))
print('Identified (4) clusters=', len(id4.groupby('Cluster_ID')))
print('Identified (5) clusters=', len(id5.groupby('Cluster_ID')))
print('Identified (6) clusters=', len(id6.groupby('Cluster_ID')))
```

```python
#save the files
id2.to_excel('allp25_matrix/fully_id_clust(2).xlsx')
id3.to_excel('allp25_matrix/fully_id_clust(3).xlsx')
id4.to_excel('allp25_matrix/fully_id_clust(4).xlsx')
id5.to_excel('allp25_matrix/fully_id_clust(5).xlsx')
id6.to_excel('allp25_matrix/fully_id_clust(6).xlsx')
```
