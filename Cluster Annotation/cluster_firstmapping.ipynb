```python
psmfiles = os.listdir('PSMTables')

print(psmfiles)

#merging the files into 1 
all_psm = []
for file in psmfiles:
    df = (
        pd.read_csv(f'PSMTables/{file}',
                     delimiter="\t")
                     .assign(File = file.split("_")[0]) #function to add a column at the end with a shorter name
    )
    all_psm.append(df) #this is to add files together

def get_spectrum(row): #function to rename the spectrum id in the psm file dataframe
    spectra_raw = row.Spectrum
    return spectra_raw.split("_")[-1]

main_psm = (pd.concat(all_psm).assign(Spectrum_ID = lambda df: df.apply(get_spectrum, axis=1))
)

main_psm
```

```python
p30files = os.listdir('C:/Users/Izadora/Documents/Project2/spectralclustering/p30_clusterfiles/rawclusters')

all_p30 = []

for file in p30files:
    df = (
        pd.read_csv(f'C:/Users/Izadora/Documents/Project2/spectralclustering/p30_clusterfiles/rawclusters/{file}', index_col=0)
    )
    all_p30.append(df)
```

```python
filenames = {'AB15.mgf':'AB15',
             'AB16.mgf':'AB16',
             'AB18.mgf':'AB18',
             'AB19.mgf':'AB19',
             'ZK01.mgf':'ZK01',
             'MM01.mgf':'MM01'}

allp30 = (
    pd.concat(all_p30)
    .sort_values('Number of Files', ascending =False)
    .rename(columns = {'Number of Files':'FileCount', 'Filename':'File'})
    .replace({'File': filenames})

)
```

```python
allp30['File'] = allp30['File'].astype(str)
allp30['Spectrum_ID'] = allp30['Spectrum_ID'].astype(int)

#this will be the working maracluster df
allp30
```

```python
#now to categorise the clusters to subcategories

print(merge.columns)

#lets clean up and only choose relevant details

clean_merge = merge[['Cluster_ID','Peptide','Type','PEPscore',
                            'FDR','Delta','Peaks_Peptide','SCORE (%)','ALC (%)',
                            'FileCount','Spectrum_ID','File','Cluster_Size']]

#now let's work with just this
clean_merge
```

```python
#let's check how many clusters and novel peptides found there are in 4-6 files

cluster_count = clean_merge['Cluster_ID'].nunique()

print('Number of filtered clusters for p30 in 4-6 files:', cluster_count)


#check the number of peptides in this clusters
peptide_count = clean_merge[['Cluster_ID','Peptide','Type']].drop_duplicates()

peptide_count = peptide_count['Type'].value_counts()

print('Type of peptides:',peptide_count)
```

```python
#now let's categorise into fully identified and unidentified
#first maybe we can rename the peptides into identified and unidentified?

group = clean_merge.groupby(by='Cluster_ID').count().reset_index()

group
```

```python
#now we can say those with clusters of 0 peptides as fully unid
unid = group.loc[group['Peptide']==0]

unid_clusters = unid[['Cluster_ID']]

print(unid_clusters)

unid_file = pd.merge(
    clean_merge,
    unid_clusters,
    how = 'right',
    on = 'Cluster_ID'
)

#unid_file.to_excel('p30_fully_unidentified.xlsx')
#unid_file.to_csv('fully_unidentified_p30.csv')

print('unidentified clusters = ', len(unid_clusters))
```

```python
#to get rid of all the clusters that are fully unidentified

filter_unid = clean_merge[~clean_merge.Cluster_ID.isin(unid_file.Cluster_ID)]

filter_unid
```

```python
#now let's group the number of peptides per cluster

peptide_filter = filter_unid.groupby(by='Cluster_ID').nunique().reset_index()


#for fully identified clusters
id = peptide_filter.loc[peptide_filter['Peptide']==1] #this would mean all the clusters that only have 1 peptides, but this could also mean blank+1 peptide (partial)

id1 = id[['Cluster_ID']]
```
