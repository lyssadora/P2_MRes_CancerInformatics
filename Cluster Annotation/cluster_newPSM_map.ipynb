Now what if we just take the cluster ID, size, file, spectrum ID, category and see what we get?

```python
p30cluster = new_p30[['Cluster_ID','FileCount','Spectrum_ID','File','Cluster_Size','Category']]

p30cluster = p30cluster.rename(columns = {'Category':'Old_Category'})
```

```python
#now we match with old and new DB

merge5 = pd.merge(
    p30cluster,
    PSM_DB,
    how = 'left',
    on = ['Spectrum_ID','File']
)

merge6 = pd.merge(
    p30cluster,
    PSM_newDB,
    how = 'left',
    on = ['Spectrum_ID','File']
)
```

Now, lets check which clusters are in which categories

```python
g1 = merge5.groupby('Cluster_ID').nunique().reset_index()

g2 = merge6.groupby('Cluster_ID').nunique().reset_index()
```

Clustering categories in this exact order:
- unidentified
- mismatch
- partial
- fully id

```python
#FOR UNIDENTIFIED CLUSTERS
unid = g1.loc[g1['Peptide']==0]

un = merge5[merge5.Cluster_ID.isin(unid.Cluster_ID)]

un['Category'] = 'Unidentified'



unid1 = g2.loc[g1['Peptide']==0]

un1 = merge6[merge6.Cluster_ID.isin(unid1.Cluster_ID)]

un1['Category'] = 'Unidentified'
```

```python
#filter out the unid from the main

filter1 = merge5[~merge5.Cluster_ID.isin(unid.Cluster_ID)]
filter2 = merge6[~merge6.Cluster_ID.isin(unid1.Cluster_ID)]
```

```python
#filter out the mismatch, where pep >1

mis1 = g1.loc[g1['Peptide']>1]
mis2 = g2.loc[g2['Peptide']>1]

#extract from the original

m1 = filter1[filter1.Cluster_ID.isin(mis1.Cluster_ID)]
m2 = filter2[filter2.Cluster_ID.isin(mis2.Cluster_ID)]

m1['Category'] = 'Mismatch'
m2['Category'] = 'Mismatch'
```

```python
#now for the partially identified
id_par1 = filter1[~filter1.Cluster_ID.isin(mis1.Cluster_ID)]
id_par2 = filter2[~filter2.Cluster_ID.isin(mis2.Cluster_ID)]

id_par1
```

```python
#for partial, it will have the isna peptides in the clusters

parc_1 = id_par1[id_par1['Peptide'].isna()]
parc_2 = id_par2[id_par2['Peptide'].isna()]

#now add to the main
par1 = id_par1[id_par1.Cluster_ID.isin(parc_1.Cluster_ID)]
par2 = id_par2[id_par2.Cluster_ID.isin(parc_2.Cluster_ID)]

par1['Category'] = 'Partial'
par2['Category'] = 'Partial'
```

```python
#for fully identified clusters, should be all else than the partial
id1 = id_par1[~id_par1.Cluster_ID.isin(parc_1.Cluster_ID)]
id2 = id_par2[~id_par2.Cluster_ID.isin(parc_2.Cluster_ID)]

id1['Category'] = 'Fully ID'
id2['Category'] = 'Fully ID'
```

Now that we have all of the clusters, we can make a new df, and save that as our new results after openMS research!

```python
new_df1 = pd.concat([id1,m1,par1,un])
new_df2 = pd.concat([id2,m2,par2,un1])
```
