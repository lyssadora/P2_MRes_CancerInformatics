```python
import os
import pandas as pd
import numpy as np
```

```python
p30 = pd.read_csv('all_p30_PSMPEAKS.csv', index_col=0)
p30
```

```python
clean = p30[['Cluster_ID','Cluster_Size','Peptide','Type','PEPscore',
                'FDR','Delta','Conflict','Annotation', 'Database','Charge','Proteins',
                'Peaks_Peptide','SCORE (%)','ALC (%)','m/z', 'z', 'RT', 'Delta RT',
                'FileCount','Spectrum_ID','File',]]

clean
```

Now we wanna categorise them, and then we can put it into the original df, and make a matrix.

Grouping identified, partial, mismatch and unidentified


```python
group = clean.groupby('Cluster_ID').count().reset_index()
```

```python
group
```

```python
#for unidentified

unid = group.loc[group['Peptide']==0]
un = clean[clean.Cluster_ID.isin(unid.Cluster_ID)]

un['Category'] = 'Unidentified'
```

```python
#filter out the unidentified from the main

filter = clean[~clean.Cluster_ID.isin(unid.Cluster_ID)]
filter
```

```python
#filter according by peptide counts
pep_fil = filter.groupby(by='Cluster_ID').nunique().reset_index()
pep_fil
```

```python
#filter out the mismatchs, pep >1

mismatch = pep_fil.loc[pep_fil['Peptide']>1]

#extract from the original

mis = clean[clean.Cluster_ID.isin(mismatch.Cluster_ID)]

mis['Category'] = 'Mismatch'
```

```python
#now for fully id and par
#filter out the mismatch ones first

id_par = pep_fil[~pep_fil.Cluster_ID.isin(mismatch.Cluster_ID)]
id_par
```

```python
filter2 = clean[clean.Cluster_ID.isin(id_par.Cluster_ID)]

filter2
```

```python
#for partial, it will have the isna peptides in the clusters

par_clust = filter2[filter2['Peptide'].isna()]
par_clust

#now add it to the main again
par = filter2[filter2.Cluster_ID.isin(par_clust.Cluster_ID)]
par['Category'] = 'Partial'
```

```python
#now for the fully id, it will all else than the par
id = filter2[~filter2.Cluster_ID.isin(par_clust.Cluster_ID)]
id['Category'] = 'Fully ID'
```

```python
#creating a new dataframe

new_df = pd.concat([id,un,par,mis])
new_df.to_csv('labelled_p30categories.csv')
```
