```python
pnovo = (
    pd.read_excel('results.xlsx')
    .rename(columns = {'Column3':'Spectrum_title',
                        'Column5':'candidate_result',
                        'Column6': 'Sequence',
                        'Column7':'Final_score',
                        'Column8':'Mod.Abundance',
                        'Column9':'Precursor_massdev',
                        'Column10':'Path rank in spectrum graph',
                        'Column11':'Score of main ions',
                        'Column12':'Score of internal ions',
                        'Column13':'Cont. score b ions',
                        'Column14':'Cont. score y ions',
                        'Column15':'Enzyme score',
                        'Column16':'Std frag. mass deviations',
                        'Column17':'max frag. mass deviations',
                        'Column18':'Charge feature',
                        'Column19':'raw score of psm',
                        'Column20':'spearman corr.',
                        'Column21':'gap feature',
                        'Column22':'gap feature for N-term'
                        }
                        )
)

pnovo = pnovo.rename (columns = {'Column4':'Spectrum Mass',
                                 'Spectrum_title': 'Spectrum'})


pnovo = pnovo.drop(columns = ['Column1', 'Column2'])

pnovo
```
