```python
import os
import pandas as pd
import numpy as np
```

```python
#work with p30_5 first

iden = pd.read_excel(
    'p30_5_id.xlsx',
    index_col=0
)

unid = pd.read_excel(
    'p30_5_unid.xlsx',
    index_col=0
)

par = pd.read_excel(
    'p30_5_partialclust.xlsx',
    index_col=0
)

mis = pd.read_excel(
    'p30_5_mismatchclust.xlsx',
    index_col=0
)
```

```python
iden
```

```python
#load the peaks files

deepnovo = pd.read_csv(
    'alldeepnovo.csv'
)

peptide = pd.read_csv(
    'allpeptide.csv'
)

psms = pd.read_csv(
    'allpsms.csv'
).drop(columns='Unnamed: 0')
```

```python
#merging them for psms

A = psms
B = unid[['Spectrum_ID','File','Cluster_ID']]
C = par[['Spectrum_ID','File','Cluster_ID']]
D = mis[['Spectrum_ID','File','Cluster_ID']]
E = iden[['Spectrum_ID','File','Cluster_ID']]

unid_psms = pd.merge(
    A,
    B,
    how = 'right',
    on = ['Spectrum_ID', 'File']
)

par_psms = pd.merge(
    A,
    C,
    how = 'right',
    on = ['Spectrum_ID', 'File']
)

mis_psms = pd.merge(
    A,
    D,
    how = 'right',
    on = ['Spectrum_ID', 'File']
)

iden_psms = pd.merge(
    A,
    E,
    how = 'right',
    on = ['Spectrum_ID', 'File']
)
```

```python
iden_psms.to_excel('iden_p30_5.xlsx')
mis_psms.to_excel('mismatch_p30_5.xlsx')
par_psms.to_excel('partial_p30_5.xlsx')
unid_psms.to_excel('unid_p30_5.xlsx')
```
