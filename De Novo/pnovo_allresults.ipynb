```python
import pandas as pd
import os
```

```python
pnovo = (
    pd.read_csv('pNovo_allspectra_results_reformatted.txt',
                header = None,
                sep = '\t')
    .drop(columns=1)
    .rename(columns = {0:'FileOrigin',
                       2:'Spectrum',
                       3:'Spectrum Mass',
                        4:'candidate_result',
                        5: 'Sequence',
                        6:'Final_score',
                        7:'Mod.Abundance',
                        8:'Precursor_massdev',
                        9:'Path rank in spectrum graph',
                        10:'Score of main ions',
                        11:'Score of internal ions',
                        12:'Cont. score b ions',
                        13:'Cont. score y ions',
                        14:'Enzyme score',
                        15:'Std frag. mass deviations',
                        16:'max frag. mass deviations',
                        17:'Charge feature',
                        18:'raw score of psm',
                        19:'spearman corr.',
                        20:'gap feature',
                        21:'gap feature for N-term'
                        }
                        )
)

```

```python
#extract the list of peptides from pnovo

peptidelist = pnovo[['Sequence']].drop_duplicates()

peptidelist.to_csv('pnovo_peptidelist.txt')
peptidelist.to_csv('pnovo_peptidelist.csv')
```

```python
def get_spectrum(row): #function to rename the spectrum id in the psm file dataframe
    spectra_raw = row.Spectrum
    return spectra_raw.split("_")[0]

def get_filename(row): #function to rename the spectrum id in the psm file dataframe
    spectra_raw = row.Spectrum
    return spectra_raw.split("_")[-1]

main_pnovo = (
    pnovo.assign(Spectrum_ID = lambda df: df.apply(get_spectrum, axis=1))
)
```

```python
main_pnovo = main_pnovo.assign(File = lambda df: df.apply(get_filename, axis=1))
```

```python
main_pnovo.head(10)
```

```python
pnovo_p1 = main_pnovo.loc[main_pnovo['candidate_result']=='P1']
```

```python
pnovo_p1.head(10)
```

```python
p30_id = pd.read_excel('p30_fully_identified.xlsx', index_col=0)
```

```python
pnovo_p1['Spectrum_ID'] = pnovo_p1['Spectrum_ID'].astype(int)
pnovo_p1['File'] = pnovo_p1['File'].astype(str)
```

```python
#merge

A = p30_id
B = pnovo_p1

merge = pd.merge(
    A,
    B,
    how = 'left',
    on = ['File','Spectrum_ID']
)
```

```python
merge.columns
```

```python
clean = merge[['Cluster_ID', 'Peptide', 'Type', 'PEPscore', 'FDR', 'Delta',
       'Peaks_Peptide', 'SCORE (%)', 'ALC (%)', 'FileCount', 'Spectrum_ID',
       'File', 'Cluster_Size', 'FileOrigin', 'Spectrum Mass', 'Sequence', 'Final_score', 'Mod.Abundance',]]

clean.to_excel('pnovo_id.xlsx')
```
