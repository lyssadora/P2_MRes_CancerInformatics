```python
import os
import pandas as pd
```

```python
peak_denovo = pd.read_csv('alldeepnovo.csv', index_col=0)
peak_denovo
```

```python
peak_psms = pd.read_csv('allpsms.csv', index_col=0)

peaks_peptide = pd.read_csv('allpeptide.csv',index_col= 0)
```

```python
peaks_peptide
```

```python
merge_peaks = [peaks_peptide, peak_psms, peak_denovo]

peaks = pd.concat(merge_peaks)
```

```python
peaks.columns
```

```python
peaks_clean = peaks[['Peptide','-10LgP', 'Mass', 'Tag Length', 'CAA (%)', 'Length', 'ppm',
       'm/z', 'z', 'RT', 'Delta RT', 'MS2 Correlation', 'Spectrum_ID', 'File','PTM', 'AScore',
       'Positional Confidence', 'tag(>=0.0%)', 'Accession', 'Found By','Area',
       'Intensity', 'Feature Id', 'Tag length', 'SCORE (%)', 'ALC (%)',
       'local confidence (%)', 'mode']]


peaks_clean
```

```python
p30_id = pd.read_excel('p30_fully_identified.xlsx', index_col=0)
p30_id
```

```python
#matching between cluster and deepnovo
A = p30_id
B = peaks_clean

merge = pd.merge(
    A,
    B,
    how = 'left',
    on =['Spectrum_ID','File']
)
```

```python
merge
```

```python
merge.columns
```

```python
clean = merge[['Cluster_ID','FileCount', 'Spectrum_ID',
       'File', 'Cluster_Size','Peptide_x','Type','PEPscore','FDR', 'Delta','Peptide_y','SCORE (%)_y','-10LgP', 'Mass', 'Tag Length',
       'CAA (%)', 'Length', 'ppm', 'm/z', 'z', 'RT', 'Delta RT','Found By']]

clean.to_excel('peaks_id.xlsx')
```

```python
p30_mis = pd.read_excel('p30_mismatchclusters.xlsx', index_col=0)

p30_mis
C = p30_mis
B = peaks_clean

merge1 = pd.merge(
    C,
    B,
    how = 'left',
    on =['Spectrum_ID','File']
)

clean1 = merge1[['Cluster_ID','FileCount', 'Spectrum_ID',
       'File', 'Cluster_Size','Peptide_x','Type','PEPscore','FDR', 'Delta','Peptide_y','SCORE (%)_y','-10LgP', 'Mass', 'Tag Length',
       'CAA (%)', 'Length', 'ppm', 'm/z', 'z', 'RT', 'Delta RT','Found By']]

clean1.to_excel('peaks_mis.xlsx')
```

```python
p30_unid = pd.read_excel('p30_fully_unidentified.xlsx', index_col=0)

D = p30_unid
B = peaks_clean

merge2 = pd.merge(
    D,
    B,
    how = 'left',
    on =['Spectrum_ID','File']
)

clean2 = merge2[['Cluster_ID','FileCount', 'Spectrum_ID',
       'File', 'Cluster_Size','Peptide_x','Type','PEPscore','FDR', 'Delta','Peptide_y','SCORE (%)_y','-10LgP', 'Mass', 'Tag Length',
       'CAA (%)', 'Length', 'ppm', 'm/z', 'z', 'RT', 'Delta RT','Found By']]


clean2.to_excel('peaks_unid.xlsx')
```
